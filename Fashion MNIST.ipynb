```python
import pandas as pd
import numpy as np

train_path = 'C:\\Users\\Rohith\\Documents\\Rohith_Stuff\\Datasets\\Fashion_MNIST\\fashion-mnist_train.csv'
test_path = 'C:\\Users\\Rohith\\Documents\\Rohith_Stuff\\Datasets\\Fashion_MNIST\\fashion-mnist_test.csv'

df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)
```

```python
train_data = df_train.values.tolist()
#train_data.remove(train_data[0])
x_train = []
y_train = []
for row in train_data:
    temp = []
    img = []
    y_train.append(row[0])
    for i in range(1,len(row)):
        if((i)%28==0):
            temp.append(row[i])
            img.append(temp)
            temp = []
        else:
            temp.append(row[i])
    x_train.append(img)
            
```

```python
test_data = df_test.values.tolist()
#train_data.remove(train_data[0])
x_test = []
y_test = []
for row in test_data:
    temp = []
    img = []
    y_test.append(row[0])
    for i in range(1,len(row)):
        if((i)%28==0):
            temp.append(row[i])
            img.append(temp)
            temp = []
        else:
            temp.append(row[i])
    x_test.append(img)
```

```python
t = []
for i in range(len(y_train)):
    q = []
    for j in range(10):
        q.append(0)
    q[y_train[i]] = 1
    t.append(q)
y_train = []
y_train = t
```

```python
t = []
for i in range(len(y_test)):
    q = []
    for j in range(10):
        q.append(0)
    q[y_test[i]] = 1
    t.append(q)
y_test = []
y_test = t
```

```python
import keras
import numpy as np
from keras.layers import Conv2D,MaxPooling2D,Dense,Dropout,Flatten
from keras.models import Sequential
from keras.layers.normalization import BatchNormalization

x_train = np.array(x_train)
y_train = np.array(y_train)

x_test = np.array(x_test)
y_test = np.array(y_test)

x_train = x_train.reshape((60000,28,28,1))


model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(28,28,1)))
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2),strides=(2,2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
#model.add(Conv2D(64,(3,3),activation='relu'))
model.add(Conv2D(128,(3,3),activation='relu'))
#model.add(Conv2D(128,(3,3),activation='relu'))
#model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(BatchNormalization())
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
             optimizer=keras.optimizers.SGD(lr=0.01),
             metrics = ['accuracy'])

model.fit(x_train,y_train,batch_size = 50,epochs=40,validation_split=0.0,verbose=1)

model.save('C:\\Users\\Rohith\\Documents\\Rohith_Stuff\\Datasets\\Fashion_MNIST\\fashion_mnist.h5')
del model
```

```python
from keras.models import load_model

model = load_model('C:\\Users\\Rohith\\Documents\\Rohith_Stuff\\Datasets\\Fashion_MNIST\\fashion_mnist.h5')
x_test = x_test.reshape((10000,28,28,1))
scores = model.evaluate(x_test,y_test,verbose=1)
print("loss:", scores[0])
print("accuracy:",scores[1])
```
